```python
# ✅ Importă bibliotecile necesare
import os
import pandas as pd
import torch
from google.colab import files
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, TrainingArguments, Trainer
from datasets import Dataset

# ✅ Verifică dacă fișierul există deja
file_path = "/content/train_dataset.csv"
if not os.path.exists(file_path):
    print("🔄 Încărcați fișierul CSV manual...")
    uploaded = files.upload()  # Selectează fișierul de pe PC

    # Mută fișierul în /mnt/data/ pentru a evita probleme la repornirea sesiunii
    for filename in uploaded.keys():
        os.rename(filename, file_path)
    print(f"✅ Fișierul a fost mutat în: {file_path}")

# ✅ Încarcă dataset-ul
df = pd.read_csv(file_path)

# ✅ Selectează primele două coloane ale dataset-ului (indiferent de nume)
if df.shape[1] < 2:
    raise ValueError("⚠️ Dataset-ul trebuie să aibă cel puțin două coloane!")

col1, col2 = df.columns[:2]  # Selectează primele două coloane
df = df[[col1, col2]]  # Păstrează doar aceste coloane
df.columns = ["input_text", "target_text"]  # Redenumește pentru compatibilitate

# ✅ Selectează 25% din dataset
df = df.sample(frac=0.25, random_state=42).reset_index(drop=True)

# ✅ Convertește într-un format compatibil cu Hugging Face Datasets
dataset = Dataset.from_pandas(df)

# ✅ Alegem un model specializat pe domeniul medical
model_name = "google/flan-t5-large"  # Model mai performant pentru text medical
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# ✅ Funcția de tokenizare
def tokenize_function(examples):
    inputs = [str(text) for text in examples["input_text"]]
    targets = [str(text) for text in examples["target_text"]]

    model_inputs = tokenizer(inputs, padding="max_length", truncation=True, max_length=256)
    labels = tokenizer(targets, padding="max_length", truncation=True, max_length=256)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# ✅ Aplică tokenizarea dataset-ului
tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["input_text", "target_text"])

# ✅ Împarte în set de antrenare și testare (80% train, 20% test)
train_test_split = tokenized_dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = train_test_split["train"]
test_dataset = train_test_split["test"]

# ✅ Definim argumentele de antrenare
training_args = TrainingArguments(
    output_dir="./medical_flan_t5",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=10,
    save_strategy="epoch",
    evaluation_strategy="epoch",
    optim="adamw_torch_fused",
    gradient_accumulation_steps=4,
    logging_steps=10,
)

# ✅ Inițializarea antrenorului Hugging Face
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)

# ✅ Antrenarea modelului
trainer.train()

# ✅ Salvează modelul antrenat
output_model_dir = "./medical_flan_t5"
trainer.save_model(output_model_dir)
tokenizer.save_pretrained(output_model_dir)

print("✅ Modelul a fost salvat în:", output_model_dir)

# ✅ Verifică dispozitivul disponibil
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)  # Mută modelul pe dispozitiv

def generate_answer(question):
    inputs = tokenizer(question, return_tensors="pt").to(device)  # Mută și input-ul pe același dispozitiv
    outputs = model.generate(**inputs, max_length=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# ✅ Testează modelul
test_input = df["input_text"].iloc[0]  # Folosește primul input din dataset
generated_output = generate_answer(test_input)
print(f"❓ Întrebare: {test_input}")
print(f"📝 Răspuns generat: {generated_output}")
```

```python
import shutil
import os

# Definirea căilor
model_source = "/content/medical_flan_t5"
model_clean = "/content/medical_flan_t5_clean"

# Creează folderul nou
os.makedirs(model_clean, exist_ok=True)
# Lista fișierelor esențiale
important_files = [
    "config.json",
    "generation_config.json",
    "model.safetensors",
    "special_tokens_map.json",
    "spiece.model",
    "tokenizer_config.json",
    "tokenizer.json",
    "training_args.bin"
]

# Copiază fișierele
for file in important_files:
    src = os.path.join(model_source, file)
    dst = os.path.join(model_clean, file)
    if os.path.exists(src):
        shutil.copy(src, dst)

print("Modelul curat a fost creat în:", model_clean)
shutil.make_archive(model_clean, 'zip', model_clean)
print("Modelul curat a fost arhivat ca:", model_clean + ".zip")
```
